```python
legit_sample=legit.sample(n=492)
new_dataset=pd.concat([legit_sample,fraud],axis =0) # axis =0 means data is added row vise if axis=1value is added column vise
```

```python
new_dataset
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# spiliting data into features and target
# feature jo h ke v1 se 28 tk or time wagera h or target me class h 0 ya 1
x = new_dataset.drop(columns='Class' ,axis=1)
y=new_dataset['Class']
```

```python
x
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, stratify=y, random_state=2)
```

```python
# x_train contains all the value of this 30 column 
# and the corresponding label which is obtain from y will be stored in y train
# 80% of data is stored in x train and corresponding label should be stored in y train
# 20% data is stored in x test and the corresponding label should be stored in y test
# if we don't mention stratify=y so distribution of 0 and 1 can be very differnt in training data and testing data
# so, if we write stratify=y there will evenly distributed 2 classes in x test and y test
# random_state=2 it means how do you want to split your data, it is nothing just to reproduce the code as it is
```

```python
print(x_train.shape, x_test.shape)
```

```python
model=LogisticRegression(solver='liblinear',max_iter=1000)
# training the logistics regression model with training data
model.fit(x_train, y_train)
```

```python
# max_iter=1000) means maximum number of iteration this solver will perform during traning
# LIBLINEAR is a linear classifier for data with millions of features 
# fit - this will fit our data to logistics model
```

```python
# now checking the performance of the model by checking the accuracy based on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction, y_train)
```

```python
# x_train_prediction me jo x-train predict kiya wo sare labels ka value h
# uske baad x_train_predictionke value ko original value se compare kr liya gya
training_data_accuracy
```

```python
# calculating accuracy of x_test
x_test_prediction = model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction, y_test)
testing_data_accuracy
```
